## Building Dimension

```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.art3d import Poly3DCollection

# Initialize lists for storing heights and widths in XZ (Wxz) and YZ (Wyz) planes
heights = []
Wyz = []  # Width in YZ plane
Wxz = []  # Width in XZ plane

# Ask for the number of floors and Ds
num_floors = int(input("Enter the number of floors: "))
Ds = int(input("Enter the number of Ds: "))
Roof_Ridge = int(input("Enter the slope of roof ridge: "))

# Loop through each floor to input its height and widths in XZ and YZ planes
for floor in range(1, num_floors + 1):
    while True:
        try:
            height = float(input(f"Enter the height for floor {floor}: "))
            width_xz = float(input(f"Enter the width in XZ plane (Wxz) for floor {floor}: "))
            width_yz = float(input(f"Enter the width in YZ plane (Wyz) for floor {floor}: "))
            
            heights.append(height)
            Wxz.append(width_xz)  # Store in Wxz list
            Wyz.append(width_yz)  # Store in Wyz list
            break
        except ValueError:
            print("Invalid input. Please enter numeric values.")

# Create a DataFrame with Floor Number as the index
df = pd.DataFrame({
    "Height (m.)": heights,
    "Width XZ (Wxz; m.)": Wxz,
    "Width YZ (Wyz; m.)": Wyz
}, index=range(1, num_floors + 1))  # Floor numbers as the index

df.index.name = "Floor Number"

# Calculate Total Height (H)
H = sum(heights)
W = Wxz + Wyz

# 3D Visualization
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot each floor as a rectangle in 3D
current_height = 0
for i, (height, width_xz, width_yz) in enumerate(zip(heights, Wxz, Wyz)):
    # Define vertices of the rectangle (XZ and YZ planes)
    vertices = [
        [0, 0, current_height],
        [width_xz, 0, current_height],
        [width_xz, width_yz, current_height],
        [0, width_yz, current_height],
        [0, 0, current_height + height],
        [width_xz, 0, current_height + height],
        [width_xz, width_yz, current_height + height],
        [0, width_yz, current_height + height]
    ]
    
    # Define faces of the 3D rectangle
    faces = [
        [vertices[0], vertices[1], vertices[5], vertices[4]],  # Bottom
        [vertices[2], vertices[3], vertices[7], vertices[6]],  # Top
        [vertices[0], vertices[4], vertices[7], vertices[3]],  # Left
        [vertices[1], vertices[5], vertices[6], vertices[2]],  # Right
        [vertices[0], vertices[3], vertices[2], vertices[1]],  # Front
        [vertices[4], vertices[5], vertices[6], vertices[7]]   # Back
    ]
    
    # Add the faces as a 3D polygon
    ax.add_collection3d(Poly3DCollection(faces, alpha=0.6, edgecolor='k'))
    
    # Update the current height
    current_height += height

# Set axis labels
ax.set_xlabel("Width in XZ (Wxz)")
ax.set_ylabel("Width in YZ (Wyz)")
ax.set_zlabel("Height (meters)")

# Set title
ax.set_title("3D Visualization of Building Dimensions")

# Show plot
plt.show()

# Display the DataFrame
print("\nBuilding Dimensions DataFrame:")
print(df)

# Display additional calculated information
print(f"\nTotal Height (H): {H} meters")
print(f"Widths in XZ plane (Wxz): {Wxz}")
print(f"Widths in YZ plane (Wyz): {Wyz}")
print(f"Value of Ds: {Ds}")
print(f"Value of W: {W}")
```

## Recommand method Use For Primary Structure

```python
# Initialize global variables
PS = 0
SS = 0
PSB = 0  # Classifier for Low or High Building

def calculate_W_over_H(H, W):
    """
    Calculate the H/W ratio for each value in W.
    """
    H_W = [H / W for W in W]
    H_W = min(H_W)  # Get the minimum H/W ratio
    return H_W

def calculate_H_over_Ds(H, Ds):
    """
    Calculate the ratio of H/Ds.
    """
    H_Ds = H / Ds
    return H_Ds

# Call the function and store the result in W_H
W_H = calculate_W_over_H(H, W)

## Case 1: Equivalent Static Wind Load (Simple Method)
def classify_Primary_Structure(H, W, Ds, H_W):
    global PS  # Use global PS variable
    if H > 80:
        PS += 1  # Increment PS if height is greater than 80
    elif H_W > 3:
        PS += 1  # Increment PS if minimum H/W ratio is greater than 3

    if PS == 0:
        CPS = "Equivalent Static Wind Load (Simple Method)"
    else:
        CPS = "Equivalent Static Wind Load (Detailed Method)"
    
    return PS, CPS

# Classify the structure
PS, CPS = classify_Primary_Structure(H, W, Ds, H_Ds)

# Classify Low Building and High Building
def classify_Class_Pristructure(H, Ds, H_W):
    global PSB  # Use global PSB variable
    # Checking for Low Building and High Building
    H_Ds = calculate_H_over_Ds(H, Ds)  # Calculate H/Ds inside the function

    if H <= 23 and H_Ds < 1:
        PSB = PSB  # If it is a Low Building, do nothing
        CPSB = "Low Building"
    else:
        PSB = PSB + 1  # Otherwise, classify as High Building
        CPSB = "High Building"
    
    return PSB, CPSB

# Classify the building class
PSB, CPSB = classify_Class_Pristructure(H, Ds, H_Ds)

#Display Parameter
print(f"H: {H}")
print(f"Ds: {Ds}")
print(f"W: {W}")
print(f"H/Ds: {H_Ds}")
print(f"W/H: {W_H}")

# Display the Method
print(f"PS: {PS}")
print(f"CPS: {CPS}")

# Display the results for Low or High Building classification
print(f"PSB: {PSB}")
print(f"CPSB: {CPSB}")
```
